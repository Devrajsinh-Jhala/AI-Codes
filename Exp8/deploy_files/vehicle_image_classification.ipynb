```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import random
import os
from keras.utils import img_to_array, array_to_img, to_categorical
from keras.layers import Conv2D, MaxPool2D, Dense, Flatten
from keras.optimizers import Adam
import cv2
```

```python
vehicle = "../../dataset/data/vehicles/"
non_vehicle = "../../dataset/data/non-vehicles/"
```

```python
def convert_image_array(image_path):
  try:
    image=cv2.imread(image_path)
    if image is not None:
      image=cv2.resize(image, (50,50))
      return (img_to_array(image))
    else:
      return np.array([])
  except Exception as e:
    print("Error", e)
```

```python
path='../../dataset/data'
root_dir=os.listdir(path)
print(root_dir)
```

```python
image_list, image_label=[],[]
temp=-1
for dir in root_dir:
  new_list = os.listdir(path+'/'+dir)
  temp+=1
  for i in new_list:
    img_path = path+'/'+dir+'/'+i 
    image_list.append(convert_image_array(img_path))
    image_label.append(temp)
print(image_label)
```

```python
norm_image_list = np.array(image_list, dtype=np.float16)/255.0
```

```python
image_label=np.array(image_label)
image_list=np.array(image_list)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(image_list, image_label, random_state=0)
```

```python
X_train=np.asarray(X_train)
y_train=np.asarray(y_train)
```

```python
len(y_train)
```

```python
from keras.models import Sequential
model = Sequential()
model.add(Conv2D(32,(3,3), padding='same', input_shape= (50,50,3),activation='relu'))
model.add(MaxPool2D(pool_size=(3,3)))
model.add(Conv2D(32,(3,3), padding='same', activation='relu'))
model.add(MaxPool2D(pool_size=(3,3)))
model.add(Conv2D(32,(3,3), padding='same', activation='relu'))
model.add(MaxPool2D(pool_size=(3,3)))
model.add(Flatten())
model.add(Dense(30,activation="relu"))
model.add(Dense(30,activation="relu"))
model.add(Dense(30,activation="relu"))
model.add(Dense(1,activation="sigmoid"))
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy', optimizer=Adam(0.001), metrics=['accuracy'])
```

```python
hist = model.fit(X_train,y_train, epochs=5, validation_split=0.2, batch_size=32)
```

```python
ypred = model.predict(X_test)
ypred1 = np.round(ypred)
ypred1 = np.asarray(ypred1, dtype='int')
```

```python
ypred1
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, ypred1)
print(cm)
print(accuracy_score(y_test, ypred1))
```
